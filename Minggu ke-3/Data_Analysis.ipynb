```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('anxiety_depression_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
# 1. Pie Chart - Distribution of Gender
plt.figure(figsize=(6,6))
# Use 'df' instead of 'data' to access the DataFrame
gender_counts = df['Gender'].value_counts()
gender_labels = gender_counts.index.tolist()
plt.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# 2. Scatter Plot - Anxiety Score vs Depression Score
plt.figure(figsize=(8,6))
# Use 'df' instead of 'data' for the DataFrame:
sns.scatterplot(x='Anxiety_Score', y='Depression_Score', data=df, hue='Gender', palette='Set1')
plt.title('Anxiety Score vs Depression Score')
plt.xlabel('Anxiety Score')
plt.ylabel('Depression Score')
plt.legend(title='Gender')
plt.show()
```

```python
# 3. Heatmap - Correlation Matrix
plt.figure(figsize=(10,8))
# Replace 'data' with 'df' to use the correct DataFrame variable:
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# 4. Box Plot - Anxiety Scores by Employment Status
plt.figure(figsize=(8,6))
# Replace 'data' with 'df':
sns.boxplot(x='Employment_Status', y='Anxiety_Score', data=df, palette='Set2')
plt.title('Anxiety Scores by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Anxiety Score')
plt.xticks([0, 1, 2, 3], ['Unemployed', 'Retired', 'Employed', 'Other'], rotation=45)
plt.show()
```
